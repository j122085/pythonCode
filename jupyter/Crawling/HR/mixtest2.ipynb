```python
import json
import os
from collections import Counter
import pandas as pd
import numpy  as np 
from  ggplot  import * 
from pylab import *

tools = ['SQL','HTML','C#','JAVA','C','PHP','JAVASCRIPT','LINUX','C++','JQUERY','PYTHON']
trans=['SQL','C#','JAVA','C','JAVASCRIPT','C++']


datadir='D:/HumanResourceData/json/'
datas={}
for filename in os.listdir(datadir):
    with open(datadir+filename) as data:
        jdata=json.load(data)
        for i in trans:
            if i not in jdata:
                jdata[i]=0
#         jdata["SQL"]=0 
#         jdata["C#"]=0
#         jdata["JAVA"]=0
#         jdata["C"]=0
#         jdata["JAVASCRIPT"]=0
#         jdata["C++"]=0
        
        for i in jdata:
            if i.lower() in ['ms sql','mysql','sql']:
                jdata["SQL"]+=jdata[i]
            if i.lower() in ['c#','c#.net']:
                jdata["C#"]+=jdata[i]
            if i.lower() in ["java"]:
                jdata["JAVA"]+=jdata[i]
            if i.lower() in ["c/c++",'c']:
                jdata["C"]+=jdata[i]
            if i.lower() in ['javascript','script','java script']:
                jdata["JAVASCRIPT"]+=jdata[i]
            if i.lower() in ["c/c++","c++"]:
                jdata["C++"]+=jdata[i]
#         print(jdata)
        datas[filename.replace(".json","")]=jdata
# print(datas)




alldata=[]


for humanResourceNet in datas:
    toolsdict={}
    toolsdict['humanResourceNet']=humanResourceNet
    for tool in datas[humanResourceNet]:
        if tool.upper() in tools:
#             print(tool.upper())
#             print(datas[humanResourceNet][tool])
            toolsdict[tool.upper()]=datas[humanResourceNet][tool]
    alldata.append(toolsdict)
alldata



import pandas as pd
df=pd.DataFrame(alldata)
x = df.groupby('humanResourceNet')
table=x.mean()
file=table.to_csv()
with open("D:/HumanResourceData/csv/all.csv","w") as f:
    f.write(file)
# 'ms sql','mysql','sql'>>sql
# 'c#','c#.net'>>c#
# "java"+"JAVA">>java
# "c/c++",'c'>>c
# 'javascript','script','java script'>>javascript
# "c/c++","c++">>c++
print(alldata)
```
