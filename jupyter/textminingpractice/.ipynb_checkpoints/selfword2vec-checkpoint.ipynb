```python
import json
with open(r'D:\Data\JsonData\TainanFood\bigtable_1.3.json') as f:
    data=json.load(f)
```

```python
text=" ".join([dien['contentcut'] for dien in data])

words=text.split(" ")
```

```python
len(words)
```

```python
appearWord=set(words)
```

```python
len(appearWord)
```

```python
wordict={word:{} for word in appearWord}
```

```python
wordict
```

```python
dim=6
n=0
windows=200

for i in range(len(words)):
    n+=1
    if n%10000==0:
        print(n)
    for j in range(i-dim,i+dim):
        try:
            if i<j:
                if words[j]+"前" not in wordict[words[i]]:
                    if len(wordict[words[i]])<windows:
                        wordict[words[i]][words[j]+"前"]=0
                wordict[words[i]][words[j]+"前"]+=1
            elif i>j:
                if words[j]+"後" not in wordict[words[i]]:
                    if len(wordict[words[i]])<windows:
                        wordict[words[i]][words[j]+"後"]=0
                wordict[words[i]][words[j]+"後"]+=1
        except:
            pass
```

```python
try:
    del wordict['']
except:
    pass
model={word:{} for word in wordict}

```

```python
import concurrent.futures
import requests

def like(wordict,word1,word2):
    publicwords=set(wordict[word1])&set(wordict[word2])
    n=0
    for word in wordict[word1]:
        n+=wordict[word1][word]**2
    m=0
    for word in wordict[word2]:
        m+=wordict[word2][word]**2
    l=0
    for word in publicwords:
        l+=wordict[word1][word]*wordict[word2][word]

    sim=l/((n**(1/2))*(m**(1/2)))
    
    return {word2:sim}

# like(wordict,'便宜','好吃')
```
