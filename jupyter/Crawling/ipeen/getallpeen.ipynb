```python
##初版
import requests
from bs4 import BeautifulSoup as bs
import os
import json
import time

def getPeenText(url):
#     url="http://www.ipeen.com.tw/comment/716052"
    res=requests.get(url)
    if res.status_code==200:
        soup=bs(res.text,'lxml')
        res.close()
        return "".join([i.text for i in soup.select('div.description')]).replace("\n","").replace("\xa0","").replace("\t","")
        
    
for filename in os.listdir("./Aipingjson/"):
    with open("./Aipingjson/"+filename) as jf:
        jsonContent=json.load(jf)
        newlist=[]
        count=0
        for olddict in jsonContent:
            time.sleep(5)
#             count+=1
#             if count%6==0:
#                 time.sleep(5)
            newdict={}
            newCommentlist=[getPeenText(url) for url in olddict["comment"]]
            olddict["comment"]=newCommentlist
            newdict=olddict
#             print(olddict)
            newlist.append(newdict)
            
        
        with open('./Aipingjson/finish' + filename, 'w') as f:
            json.dump(newlist, f)
```
