```python
for dien in bigTable:
    if dien['style']=="":
         dien['style']='其它'
    if dien['introduction']=="":
         dien['introduction']='無'
    if dien['address']=="":
         dien['address']='臺南市'
    if dien['tel']=="":
         dien['tel']='無'
    if dien['score']=="":
         dien['score']='0'
    if 'coordinate' not in dien or dien['coordinate']==None:
        dien['coordinate']={'lat':22.9982,'lng':120.203}
    if 'lat' not in dien['coordinate']:
        dien['coordinate']['lat']=22.9982
    if 'lng' not in dien['coordinate']:
        dien['coordinate']['lng']=120.203
```

```python
trans={'食物美味':'delicious', '食物很香': 'fragrant', '食物精緻漂亮':'pretty', '其他正面情緒': 'Positive',\
       '環境好': 'environment', '份量大': 'big', '偏甜': 'sweet', '划算': 'cheap', '口味清淡': 'Lighttaste',\
       '偏酸': 'acid', '服務好': 'goodserve', '小份量': 'small', '重口味': 'Heavytaste', '偏辣': 'hot','無評': 'nopeen'}

for dien in bigTable:
    dien['engfeaturescores']={}
    for feature in dien['featurescores']:
        dien['engfeaturescores'][trans[feature]]=dien['featurescores'][feature]
```
