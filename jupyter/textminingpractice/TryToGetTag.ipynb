```python
import json
bigtable=[]
for raws,scores,tfidfs in zip(rawtable,pricetable,tfidftable):
#     print(raws["name"])
    dien={}
    dien["name"]=raws["name"]
    dien["style"]=raws["style"]
    dien["wordcount"]=tfidfs["wordCount"]
    dien["contentcut"]=scores["contentcut"]
    dien["N"]=scores["n"]
    try:
        dien["score"]=scores["price"]
    except:
        dien["score"]=0
    dien["tfidf"]=tfidfs["TF_IDF"]
    bigtable.append(dien)
    
with open("D:\Data\JsonData\TainanFood\BigTable.json","w") as f:
    json.dump(bigtable,f)
```

```python
#同義字
with open("D:\Data\csvtest\MyDict.csv") as f:
    csv=f.read()
x=[{j:i.split(",")[1] for j in i.split(",")[1:]} for i in csv.split("\n") if i.split(",")[0]!=""]
pinsameword={}
for i in x:
    pinsameword.update(i)
del pinsameword[""]
```

```python
##做各種評論的分數加總

#i為每家店的json
for i in bigtable:
    featurescore={}
    #j為每個詞
    for j in i['wordcount']:
        #pinsameword為同義字字典
        if j in pinsameword:
            
            
            if not pinsameword[j] in featurescore:
                if not pinsameword[j][1:] in featurescore:
                    if pinsameword[j][0]=="否":
                        featurescore[pinsameword[j][1:]]=0
                    else:
                        featurescore[pinsameword[j]]=0
                
                
                
#                 #(分數由同義詞出現次數/評論數)*
                if pinsameword[j][0]=="否":
                    featurescore[pinsameword[j][1:]]-=round(i['wordcount'][j]/i["N"],1)*1
                else:
                    featurescore[pinsameword[j]]+=round(i['wordcount'][j]/i["N"],1)
        
    if len(featurescore)==0:
        featurescore["無評"]=1
    i['featurescore']=featurescore
    
```
