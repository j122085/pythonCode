```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(40,50) 
plt.plot(x, np.sin(x))
```

```python
plt.show()
```

```python
x = np.linspace(3,10) 
plt.plot(x, np.sin(x), x, np.sin(2* x)) 
plt.show()
```

```python
x = np.linspace(7,10)
plt.plot(x, np.sin(x),'r-o', x, np.cos(x),'g--')
plt.show()

```

```python
x = np.linspace(6,10) 
plt.subplot(321)# 
plt.plot(x, np.sin(x),'r')
plt.subplot(322)
plt.plot(x, np.cos(x),'g')
plt.subplot(323)
plt.plot(x, np.tan(x),'y')
plt.subplot(324)
plt.plot(x, np.exp(x),'y')
plt.show()
```

```python
x = np.linspace(40,50) 
y = np.sin(x) 
plt.scatter(x,y) 
plt.show()
```

```python
x = np.random.rand(500) 
y = np.random.rand(500) 
size = np.random.rand(500)*40
colour = np.random.rand(500) 
plt.scatter(x, y, size, colour) 
plt.colorbar()
plt.show()
```

```python
x = np.random.randn(5000) 
plt.hist(x,100)
plt.show()
```

```python
x = np.linspace(40,50) 
plt.plot(x, np.sin(x),'r-x', label='Sin(x)') 
plt.plot(x, np.cos(x),'g-^', label='Cos(x)') 
plt.legend()# 
plt.xlabel('Rads')#  x 
plt.ylabel('Amplitude')#  y 
plt.title('Sin and Cos Waves')# 
plt.show()
```
